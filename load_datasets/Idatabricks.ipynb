## Load databricks-dolly Dataset

```python
from datasets import load_dataset
import json

ds = load_dataset("databricks/databricks-dolly-15k", split="train")
```

```python
ds[0]
```

```python
from pathlib import Path
import json

folder = Path("../Idatabricks/")
folder.mkdir(parents=True, exist_ok=True)

json_path = folder / "chat.jsonl"

with open(json_path, "w", encoding="utf-8") as f:
    for i, data in enumerate(ds):
        # 組 user prompt
        user_text = (
            "Read the following context and answer the question.\n\n"
            f"Context: {data['context']}\n\n"
            f"Question: {data['instruction']}"
        )

        sample = {
            "id": f"text_instruct_dolly_{i:06d}",
            "task_type": "text_instruct",
            "images": [],
            "bboxes": [],
            "grounding_targets": [],
            "conversations": [
                {"from": "human", "value": user_text},
                {"from": "gpt", "value": data["response"]},
            ],
            "meta": {
                "src_dataset": "databricks-dolly-15k"
            },
        }

        f.write(json.dumps(sample, ensure_ascii=False) + "\n")

print(f"已儲存至 {json_path}")
```
