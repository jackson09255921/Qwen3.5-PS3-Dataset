```python
from datasets import load_dataset
import json

ds = load_dataset('allenai/tulu-3-sft-personas-instruction-following', split='train')
```

```python
ds[0]
```

```python
from pathlib import Path
import json

folder = Path("../tulu_personas/")
folder.mkdir(exist_ok=True, parents=True)
out_path = folder / "chat.jsonl"

with out_path.open("w", encoding="utf-8") as f:
    for idx, item in enumerate(ds):
        msgs = item["messages"]
        # 只保留標準 user→assistant 結構
        if len(msgs) < 2 or msgs[0]["role"] != "user" or msgs[1]["role"] != "assistant":
            continue

        sample = {
            "id": f"text_instruct_tulu_personas_{idx:06d}",
            "task_type": "text_instruct",
            "images": [],
            "bboxes": [],
            "grounding_targets": [],
            "conversations": [
                {"from": "human", "value": msgs[0]["content"]},
                {"from": "gpt", "value": msgs[1]["content"]},
            ],
            "meta": {
                "src_dataset": "tulu-3-sft-personas-instruction-following",
                "orig_id": item["id"],
            },
        }

        f.write(json.dumps(sample, ensure_ascii=False) + "\n")

print(f"saved to {out_path}")
```
