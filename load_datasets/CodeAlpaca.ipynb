```python
from datasets import load_dataset
import json

ds = load_dataset('HuggingFaceH4/CodeAlpaca_20K', split='train')
```

```python
ds[0]
```

```python
from pathlib import Path
import json

folder = Path('../CodeAlpaca/')
folder.mkdir(exist_ok=True, parents=True)
with open(folder/'chat.jsonl', 'w') as f:
    for idx, item in enumerate(ds):
        data = {
            "id": f"codealpaca_{idx}",
            "task_type": "text_instruct",
            "images": [],
            "bboxes": [],
            "grounding_targets": [],
            "conversations": [
                {"from": "human", "value": item["prompt"].strip()},
                {"from": "gpt", "value": item["completion"].strip()},
            ],
        }
        f.write(json.dumps(data) + '\n')

print("Finished writing chat.jsonl")
```
