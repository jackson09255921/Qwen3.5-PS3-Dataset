```python
from datasets import load_dataset
import json

ds = load_dataset("Open-Orca/FLAN", split="train", streaming=True)
ds_head = ds.take(20_000_000)   # 只取前 20M，回傳 iterator

with open("./FLAN/chat.jsonl", "w", encoding="utf-8") as f:
    for i, example in enumerate(ds_head):
        data = {
            "id": str(i),
            "image": None,
            "conversations": [
                {"from": "human", "value": example["inputs"]},
                {"from": "gpt", "value": example["targets"]},
            ],
        }
        f.write(json.dumps(data, ensure_ascii=False) + "\n")
```
