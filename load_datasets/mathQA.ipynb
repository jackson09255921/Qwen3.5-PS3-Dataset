```python
from datasets import load_dataset
import json

ds = load_dataset("allenai/math_qa", revision="refs/pr/5", split="train")
```

```python
ds[0]
```

```python
from pathlib import Path
import json

folder = Path('../math_qa/')
folder.mkdir(exist_ok=True, parents=True)
with open(folder/'chat.jsonl', 'w') as f:
    for idx, item in enumerate(ds):
        problem = item["Problem"].strip()
        options = item["options"].strip()
        correct = item["correct"].strip()
        user = (
            "You will be given a multiple-choice math word problem.\n"
            "Choose the correct option and answer with the option letter only.\n\n"
            f"Problem: {problem}\n"
            f"Options: {options}"
        )
        data = {
            "id": f"mathqa_{idx}",
            "task_type": "text_instruct",
            "images": [],
            "bboxes": [],
            "grounding_targets": [],
            "conversations": [
                {"from": "human", "value": user},
                {"from": "gpt", "value": correct},
            ],
        }
        f.write(json.dumps(data)+'\n')
```
